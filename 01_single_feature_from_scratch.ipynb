```python
import pandas as pd, numpy as np
df = pd.read_csv("House_Rent_Dataset.csv")
df.head()
```

```python
df.info()
```

```python
#Rent will be our target variable and for starters our independent variable will be size
X = df['Size'].values
X = X.reshape(-1,1)
y = df['Rent'].values
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X,y, alpha = 0.3)
plt.xlabel("Area(sq ft)")
plt.ylabel("Rent per month (in Rupees)")
plt.title("Area v/s Rent")
plt.show()
```

```python
df['Rent'].describe()
```

```python
print(f"Peak to Peak range by column in Raw        y:{np.ptp(y)}")
```

```python
low = np.percentile(df['Rent'],1)
high = np.percentile(df['Rent'],99)
low , high
```

```python
df_clean = df[(df['Rent'] >= low) & (df['Rent'] <= high)]
```

```python
X = df_clean[['Size']].values
y = df_clean['Rent'].values
X.shape, y.shape
```

```python
plt.scatter(X,y, alpha = 0.3)
plt.xlabel("Area(sq ft)")
plt.ylabel("Rent per month (in Rupees)")
plt.title("Area v/s Rent(w/o outliers)")
plt.show()
```
